```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ast
```

```python
movie = pd.read_csv('https://github.com/sukhioo7/Resources/blob/main/tmdb_5000_movies.csv?raw=true')
```

```python
credits = pd.read_csv('https://github.com/sukhioo7/Resources/blob/main/tmdb_5000_credits.csv?raw=True')
```

```python
movies = movie.merge(credits,on='title')
```

```python
movies.info()
```

```python
movies = movies[['movie_id','title','overview','keywords','genres','cast','crew']]
```

```python
movies.head()
```

```python
a = movies['genres'][0]
```

```python
a
```

```python
ast.literal_eval(a)
```

```python
def convert_genres(item):
  temp_lst = []
  item = ast.literal_eval(item)
  for dct in item:
    temp_lst.append(dct['name'])
  return temp_lst
```

```python
convert_genres(a)
```

```python
movies['genres'] = movies['genres'].apply(convert_genres)
```

```python
movies.head()
```

```python
movies['keywords'][0]
```

```python
def convert_keywords(item):
  temp_lst = []
  item = ast.literal_eval(item)
  count = 1
  for dct in item:
    if count<=6:
      temp_lst.append(dct['name'])
    else:
      break
  return temp_lst
```

```python
movies['keywords'] = movies['keywords'].apply(convert_keywords)
```

```python
movies['cast'][0]
```

```python
def convert_cast(item):
  temp_lst = []
  item = ast.literal_eval(item)
  count = 1
  for dct in item:
    if count<=3:
      temp_lst.append(dct['name'])
    else:
      break
  return temp_lst
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
def convert_crew(item):
  temp_lst = []
  item = ast.literal_eval(item)
  for dct in item:
    if dct['job']=='Director':
      temp_lst.append(dct['name'])
  return temp_lst
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.reset_index(drop=True,inplace=True)
```

```python
movies['overview'] = movies['overview'].apply(lambda item: item.split())
```

```python
def remove_space(item):
  temp_lst = []
  for i in item:
    temp_lst.append(''.join(i.split()))
  return temp_lst
```

```python
movies['keywords'] = movies['keywords'].apply(remove_space)
```

```python
movies['genres'] = movies['genres'].apply(remove_space)
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
```

```python
movies['crew'] = movies['crew'].apply(remove_space)
```

```python
movies['Tag'] = movies['overview'] +  movies['keywords'] +  movies['genres'] +  movies['cast'] +  movies['crew']
```

```python
movies.head()
```

```python
movies_2 = movies[['movie_id','title','Tag']]
```

```python
movies_2.head()
```

```python
from nltk.stem.porter import PorterStemmer

stemming = PorterStemmer()
```

```python
def stemming_words(items):
  temp_lst = []
  for word in items:
    temp_lst.append(stemming.stem(word))
  return temp_lst
```

```python
movies_2['Tag'] = movies_2['Tag'].apply(stemming_words)
```

```python
movies_2.head()
```

```python
movies_2['Tag'] = movies_2['Tag'].apply(lambda item: ' '.join(item))
```

```python
movies_2.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vector = CountVectorizer(max_features=5000,stop_words='english')
```

```python
count_vector.fit(movies_2['Tag'])
```

```python
movie_matrix = count_vector.transform(movies_2['Tag']).toarray()
```

```python
movie_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similar_matrix = cosine_similarity(movie_matrix)
```

```python
def recommend(movie):
    index = movies_2[movies_2['title'] == movie].index[0]
    distances = sorted(list(enumerate(similar_matrix[index])),
                       reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(movies_2.iloc[i[0]]['title'])
```

```python
recommend("Interstellar")
```
